#### 1. 
Предложить пример функции,  не имеющей предела в нуле и в бесконечностях.
#### ОТВЕТ:

$f(x) = \sin{x} + \sin{(1/x)}$, где $\sin{x}$ обеспечит неопределённость на $\infty$, а $\sin{(1/x)}$ - неопределённость около $0$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

x = np.linspace(0.0001, 10, 1000)
y = np.sin(x) + np.sin(1/x)
plt.plot(x, y)
```

#### 3. 
Исследовать функцию $f(x) = x^3 - x^2$ по плану

Область задания и область значений   
Нули функции и их кратность   
Отрезки знакопостоянства   
Интервалы монотонности   
Четность функции   
Ограниченность   
Периодичность   
#### ОТВЕТ:

```python
x = np.linspace(-100, 100, 1000)
y = x**3 - x**2
plt.plot(x, y)
plt.grid()
```

Область задания и область значений: $x \in ]-\infty, +\infty[, \; y \in ]-\infty, +\infty[$   
Нули функции и их кратность: $x^3 - x^2 = 0$ => нули функции $x_1 = 0, x_2 = 0, x_3 = 0$   
Отрезки знакопостоянства: при $x \in ]-\infty, 0[ \; y < 0$, при $x \in ]0,+\infty[ \; y > 0$   
Интервалы монотонности: функция __возрастает__ при $x \in ]-\infty, \infty[$     
Четность функции:  $f(x) = x^3 - x^2, \; f(-x) = -x^3 - x^2$, т.е. $f(x) \ne f(-x)$ и $-f(x) \ne f(-x)$ => это __функция общего вида__  
Ограниченность: функция __не ограничена__   
Периодичность: функция __непериодическая__  

#### 4.
 $\displaystyle \lim_{x \rightarrow 0} \frac{3x^3 - 2x^2}{4x^2}$  
#### ОТВЕТ:

 $\displaystyle \lim_{x \rightarrow 0} \frac{3x^3 - 2x^2}{4x^2} =  \lim_{x \rightarrow 0} \frac{3x}{4} - 1/2 = -1/2$   

#### 5.
 $\displaystyle \lim_{x \rightarrow 0} \frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}$  
#### ОТВЕТ:

 $\displaystyle \lim_{x \rightarrow 0} \frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1} = \lim_{y \rightarrow 1} \frac{(\sqrt{y}-1)\cdot(\sqrt{y}+1)\cdot(\sqrt[3]{y^2}+\sqrt[3]{y}+1)}{(\sqrt[3]{y}-1)\cdot(\sqrt[3]{y^2}+\sqrt[3]{y}+1)\cdot(\sqrt{y}+1)} = \lim_{y \rightarrow 1} \frac{(y-1)\cdot(\sqrt[3]{y^2}+\sqrt[3]{y}+1)}{(y-1)\cdot(\sqrt{y}+1)} = 3/2 = 1,5$  

#### 6.
 $\displaystyle \lim_{x \rightarrow \infty} \Bigl (\frac{x+3}{x} \Bigr )^{4x+1}$  
#### ОТВЕТ:

$\displaystyle \lim_{x \rightarrow \infty} \Bigl (\frac{x+3}{x} \Bigr )^{4x+1} = \lim_{4x \rightarrow \infty} \Bigl (1+\frac{12}{4x} \Bigr )^{4x+1} = e^{12}$  

#### 7.
 $\displaystyle \lim_{x \rightarrow 0} \frac{\sin(2x)}{4x}$  
#### ОТВЕТ:

 $\displaystyle \lim_{x \rightarrow 0} \frac{\sin(2x)}{4x} = \lim_{2x \rightarrow 0} \frac{\sin(2x)}{2\cdot(2x)} = 1/2$ 

#### 8.
 $\displaystyle \lim_{x \rightarrow 0} \frac{x}{\sin(x)}$  
#### ОТВЕТ:

$\displaystyle \lim_{x \rightarrow 0} \frac{x}{\sin(x)} = \frac{1}{\lim_{x \rightarrow 0}\frac{\sin(x)}{x}} = 1/1 = 1$  

#### 9.
 $\displaystyle \lim_{x \rightarrow 0} \frac{x}{\arcsin(x)}$  
#### ОТВЕТ:

 $\displaystyle \lim_{x \rightarrow 0} \frac{x}{\arcsin(x)} = \frac{1}{\lim_{x \rightarrow 0} \frac{\arcsin(x)}{x}} = 1/1 = 1$  

#### 10.
 $\displaystyle \lim_{x \rightarrow \infty} \Bigl (\frac{4x+3}{4x-3} \Bigr )^{6x}$  
#### ОТВЕТ:

$\displaystyle \lim_{x \rightarrow \infty} \Bigl (\frac{4x+3}{4x-3} \Bigr )^{6x} = \lim_{x \rightarrow \infty} \Bigl (1 + \frac{-6}{4x-3} \Bigr )^{\frac{-(4x-3)}{6}\cdot 6x\cdot\frac{-6}{4x-3}} = \exp \Bigl (\lim_{x \rightarrow \infty}\frac{-6\cdot 6x}{4x-3}\Bigr) = e^{-36/4} = e^{-9}$  

#### 11.
 $\displaystyle \lim_{x \rightarrow \infty} \frac{\sin{x}+\ln{x}}{x}$  
#### ОТВЕТ:

 $\displaystyle \lim_{x \rightarrow \infty} \frac{\sin{x}+\ln{x}}{x} = \lim_{x \rightarrow \infty} \frac{\sin{x}}{x} + \lim_{x \rightarrow \infty} \frac{\ln{x}}{x} = 0 + 0 = 0$
