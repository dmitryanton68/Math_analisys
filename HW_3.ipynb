13. 
$ f(x) = x^2$ разложить функцию в ряд Фурье по косинусам на отрезке $ x \in [-2, 0]$, построить график функции и ее разложения

$ f(x) = x^2$ чётная функция, поэтому $\displaystyle f(x) = a_0 + \sum\limits_{n=1}^\infty{a_n\:\cos(nx)}$ и при разложении на отрезке $[-\pi, \pi]$ получим   
   
$\displaystyle a_0 = \frac{1}{2\pi} \int_{-\pi}^{\pi} x^2 \,\mathrm{d}x = \frac{1}{2\pi} \cdot \frac{x^3}{3} \Bigr |_{-\pi}^{\pi} = \frac{\pi^2}{3}$   
   
$\displaystyle a_n = \frac{1}{\pi} \int_{-\pi}^{\pi} x^2 \cos{(nx)}\,\mathrm{d}x = \frac{1}{\pi} \int_{-\pi}^{\pi} x^2  \frac{1}{n}\,\mathrm{d}\sin{(nx)} =$   
$\displaystyle=\frac{1}{\pi} \cdot \Bigl(x^2 \cdot \frac{\sin{nx}}{n} \Bigr|_{-\pi}^{\pi} - \frac{1}{n}\int_{-\pi}^{\pi} \sin{(nx)}\,\mathrm{d}x^2 \Bigr) = \frac{1}{\pi} \cdot \Bigl(x^2 \cdot \frac{\sin{nx}}{n} \Bigr|_{-\pi}^{\pi} - \frac{2}{n}\int_{-\pi}^{\pi}x \sin{(nx)}\,\mathrm{d}x \Bigr) =$   
$\displaystyle= \frac{2}{\pi n} \cdot \Bigl(\frac{x}{n} \cdot \cos{(nx)} \Bigr|_{-\pi}^{\pi} - \frac{1}{n}\int_{-\pi}^{\pi} \cos{(nx)}\,\mathrm{d}x \Bigr) = \frac{2}{\pi n^2} \cdot \Bigl(x \cdot \cos{(nx)} \Bigr|_{-\pi}^{\pi} \Bigr) = \frac{2 \cdot 2\pi}{\pi n^2} \cdot(-1)^n = \frac{4}{n^2} \cdot(-1)^n$   
   
то есть $\displaystyle f(x) = a_0 + \sum\limits_{n=1}^\infty{a_n\:\cos(nx)} = \frac{\pi^2}{3} + 4 \cdot \sum\limits_{n=1}^\infty \frac{\cos{nx}}{n^2} \cdot (-1)^n$

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
def a(x, n):
    return 4*((-1)**n)*np.cos(n*x)/n**2

def f(x):
    func = x**2
    summ = np.pi**2/3
    for n in range(1, 10):
        a_i = a(x, n)
        summ += a_i
    return summ
```

```python
x = np.linspace(-2, 0, 20)
y = x**2
plt.plot(x, y, color='red')
plt.scatter(x, f(x), color='blue')
plt.grid()
```

Разложение $ f(x) = x^2$ на отрезке $[-2, 0]$ (для удобства обозначено  $L = 2$):   
   
$\displaystyle a_0 = \frac{1}{L} \int_{-L}^{0} x^2 \,\mathrm{d}x = \frac{1}{L} \cdot \frac{x^3}{3} \Bigr |_{-L}^{0} = \frac{L^2}{3} = 4/3$   
   
$\displaystyle a_n = \frac{2}{L} \int_{-L}^{0} x^2 \cos{(\frac{n\pi}{L}x)}\,\mathrm{d}x = \frac{2}{L} \int_{-L}^{0} x^2  \frac{L}{n\pi}\,\mathrm{d}\sin{(\frac{n\pi}{L}x)} =$   
$\displaystyle=\frac{2}{L} \cdot \Bigl(x^2 \cdot \frac{L}{n\pi}\sin{(\frac{n\pi}{L}x)} \Bigr|_{-L}^{0} - \frac{L}{n\pi} \int_{-L}^{0} \sin{(\frac{n\pi}{L}x)}\,\mathrm{d}x^2 \Bigr) = \frac{2}{L} \cdot \Bigl(x^2 \cdot \frac{L}{n\pi} \sin{(\frac{n\pi}{L}x)} \Bigr|_{-L}^{0} - \frac{2L}{n\pi}\int_{-L}^{0}x \sin{(\frac{n\pi}{L}x)}\,\mathrm{d}x \Bigr) =$   
$\displaystyle= -\frac{4}{n\pi} \cdot (-\frac{L}{n\pi}) \cdot \Bigl(x \cdot \cos{(\frac{n\pi}{L}x)} \Bigr|_{-L}^{0} - \int_{-L}^{0} \cos{(\frac{n\pi}{L}x)}\,\mathrm{d}x \Bigr) = \frac{4L}{n^2 \pi^2} \cdot \Bigl(x \cdot \cos{(\frac{n\pi}{L}x)} \Bigr) \Bigr|_{-L}^{0} = \frac{4 \cdot L^2}{n^2 \pi^2} \cdot(-1)^n = \frac{16}{n^2\pi^2} \cdot(-1)^n$   
   
то есть $\displaystyle f(x) = a_0 + \sum\limits_{n=1}^\infty{a_n\:\cos{(\frac{n\pi}{L}x})} = \frac{4}{3} + 16 \cdot \sum\limits_{n=1}^\infty \frac{\cos{(\frac{n\pi}{2}x)}}{n^2\pi^2} \cdot (-1)^n$

```python
def a1(x, n):
    return 16*((-1)**n)*np.cos(n*np.pi*x/2)/(n**2*np.pi**2)

def f1(x):
    func = x**2
    summ = 4/3
    for n in range(1, 10):
        a_i = a1(x, n)
        summ += a_i
    return summ
```
